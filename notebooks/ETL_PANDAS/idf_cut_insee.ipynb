```python
import pandas as pd
# On enregistre le path de l'objet (dans un bucket S3)
bucket_path = 'gs://mspr-data-ia-raw/dossier_complet.csv'

# On initie un dataframe avec Pandas
df = pd.read_csv(bucket_path, sep=';')

df.head()
```

```python
# Liste des codes départements de l’Île-de-France (SAUF PARIS)
codes_idf = ['77', '78', '91', '92', '93', '94', '95']

# Filtre pour tout enlever sauf l'idf hors paris
df_idf = df[
    (df['CODGEO'].str.len() == 5) &
    (df['CODGEO'].str[:2].isin(codes_idf))
]


# Affichage des premières lignes pour vérif
df_idf.head()
```

```python
from google.cloud import storage

client = storage.Client()
bucket = client.get_bucket('mspr-data-ia-raw')  # Nom du bucket
blob = bucket.blob('idf_INSEE_dossier_complet.csv')   # Nom du nouveau csv

# Sauvegarde locale temporaire
df_idf.to_csv('temp_export.csv', index=False)

# Upload
blob.upload_from_filename('idf_insee_temp_export.csv')
```
