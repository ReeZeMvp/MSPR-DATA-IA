```python
import pandas as pd
# On enregistre le path de l'objet (dans un bucket S3)
bucket_path = 'gs://mspr-data-ia-raw/Ministere_Interieur_deliquance.csv'

# On initie un dataframe avec Pandas
df = pd.read_csv(bucket_path, sep=';')

df.head()
```

```python
# On renomme puis dans la colonne 'CODGEO' on remplit les valeur N/A puis on converti en string
df.rename(columns={'CODGEO_2024': 'CODGEO'}, inplace=True)
df['CODGEO'] = df['CODGEO'].fillna('').astype(str)
# Liste des codes départements de l’Île-de-France (SAUF PARIS)
codes_idf = ['77', '78', '91', '92', '93', '94', '95']

# Filtre pour tout enlever sauf l'idf hors paris
df_idf = df[
    (df['CODGEO'].str.len() == 5) &
    (df['CODGEO'].str[:2].isin(codes_idf))
]

# Affichage des premières lignes pour vérif
df_idf.head()
```

```python
# Liste d'infraction à garder
categories_secu = [
    "Cambriolages de logement",
    "Destructions et dégradations volontaires",
    "Vols sans violence contre des personnes"
]

# Filtrage du DataFrame
sorted_df_idf = df_idf[df_idf["indicateur"].isin(categories_secu)]
# Affichage pour vérif
display(sorted_df_idf)
```

```python
# on constante que certains taux_pour_mille sont en NaN, et que l'on retrouve le taux dans complement_info_taux, il nous faut donc consolider cela
sorted_df_idf.loc[:, "nombre"] = sorted_df_idf["nombre"].fillna(sorted_df_idf["complement_info_nombre"])
sorted_df_idf.loc[:, "taux_pour_mille"] = sorted_df_idf["taux_pour_mille"].fillna(sorted_df_idf["complement_info_taux"])
display(sorted_df_idf)
```

```python
# on évite le data leakage en supprimant toutes les données postérieures à l'élection (2022)
sorted_df_idf = sorted_df_idf[sorted_df_idf["annee"] <= 2022]
display(sorted_df_idf)
```

```python
sorted_df_idf["taux_pour_mille"] = sorted_df_idf["taux_pour_mille"].astype(str).str.replace(",", ".", regex=False).astype(float)

sorted_df_idf = sorted_df_idf.sort_values(by=["CODGEO", "indicateur", "annee"])

df_aggr = (
    sorted_df_idf.groupby(["CODGEO", "indicateur"])
    .agg(
        taux_actuel=("taux_pour_mille", lambda x: x.iloc[-1]), # année 2022
        #taux_moyen=("taux_pour_mille", lambda x: x.tail(3).mean()),  # moyenne 2020–2022
        #taux_ecart_type=("taux_pour_mille", lambda x: x.tail(3).std()),  # volatilité récente
        #taux_min=("taux_pour_mille", lambda x: x.tail(3).min()),
        #taux_max=("taux_pour_mille", lambda x: x.tail(3).max()),
        #taux_derniere_annee=("taux_pour_mille", lambda x: x.loc[x.index[-2]]),
        taux_variation=("taux_pour_mille", lambda x: x.iloc[-1] - x.iloc[-5] if len(x) >= 3 else None)  # tendance sur 5 ans
    )
    .reset_index()
)

display(df_aggr)
```

```python
df_secu_final = (
    df_aggr.groupby("CODGEO")
    .agg(
        taux_securite_actuel=("taux_actuel", "mean"),
        taux_securite_variation=("taux_variation", "mean")
    )
    .reset_index()
)
display(df_secu_final)
```

```python
from google.cloud import bigquery
# Export avec l'API BigQuery native
def export_with_native_api(df, project_id, dataset_id, table_id):
    """
    Alternative avec l'API BigQuery native
    """

    client = bigquery.Client(project=project_id)
    full_table_id = f"{project_id}.{dataset_id}.{table_id}"

    try:
        # Configuration du job
        job_config = bigquery.LoadJobConfig(
            write_disposition="WRITE_TRUNCATE",  # Équivalent à 'replace'
            autodetect=True,  # Détection automatique du schéma
        )

        print(f"⬆️ Upload avec l'API native...")
        job = client.load_table_from_dataframe(df, full_table_id, job_config=job_config)
        job.result()  # Attendre la fin du job

        # Vérification
        table = client.get_table(full_table_id)
        print(f"✅ Upload réussi avec l'API native")
        print(f"📊 Lignes dans BigQuery : {table.num_rows}")

    except Exception as e:
        print(f"❌ Erreur avec l'API native : {str(e)}")

# Utilisation de nos données
df = df_secu_final
project_id = "mspr-data-ia"
dataset_id = "idf_silver_cleaned"
table_id = "dim_securite"

print("\n=== API BigQuery native ===")
export_with_native_api(df, project_id, dataset_id, table_id)
```
